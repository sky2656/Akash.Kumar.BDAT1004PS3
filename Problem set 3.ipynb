# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
roman_emperors = pd.read_csv('C:/Users/sky26/Downloads/BDAT/BDAT 1004 Data Programming/Assignment/Problem set 3/roman-emperor-reigns.csv')

# Filter assassinated emperors
assassinated_emperors = roman_emperors[
    roman_emperors['Cause_of_Death'].str.lower().str.contains('assassinated', na=False)]

# Print assassinated emperors
print(assassinated_emperors)

# Calculate the number of assassinated emperors and other deaths
number_assassinated = len(assassinated_emperors)
other_deaths = len(roman_emperors) - number_assassinated

# Print the counts
print("Number of assassinated emperors:", number_assassinated)
print("Number of other deaths:", other_deaths)

# Create a smaller pie chart
fig, ax = plt.subplots(figsize=(8, 8))
emperor = assassinated_emperors["Emperor"]
cause_of_death = assassinated_emperors["Cause_of_Death"]

# Plot a pie chart with labels, percentages, and other parameters
pie = ax.pie(range(len(cause_of_death)), labels=emperor, autopct='%1.2f%%', startangle=50, radius=0.045 * 100, rotatelabels=270)

# Set text in the pie chart bold
for text in pie[1]:
    text.set_fontweight('bold')

# Set plot title
plt.title('Assassinated Roman Emperors')

# Display the plot
plt.show()
```

# Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# File URL
file_url = 'C:/Users/sky26/Downloads/BDAT/BDAT 1004 Data Programming/Assignment/Problem set 3/arcade-revenue-vs-cs-doctorates.csv'

# Read data into pandas DataFrame
df = pd.read_csv(file_url)

# Rename columns for clarity
df.rename(columns={'Total Arcade Revenue (billions)': 'REVENUE', 'Computer Science Doctorates Awarded (US)': 'AWARDS'}, inplace=True)

# Group by Year
groups = df.groupby('Year')

# Plot each group as scatter plot
for name, group in groups:
    plt.scatter(group.REVENUE, group.AWARDS, label=name, marker='o', s=100)

# Set labels and legend
plt.legend()
plt.xlabel("Arcade Revenue (billions)")
plt.ylabel("Computer Science Doctorates Awarded (US)")

# Show the scatter plot
plt.show()
```
